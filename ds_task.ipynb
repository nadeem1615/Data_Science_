```python
import pandas as pd
student_info=pd.read_csv('/university_admissions_data.csv')
student_info
```

```python
input_dataset=student_info.drop(columns=['Internet_Availability','Final_Year_Result'])
input_dataset
```

```python
output_dataset=student_info[['Internet_Availability','Final_Year_Result']]
output_dataset
```

```python
from sklearn.tree import DecisionTreeClassifier
student_model=DecisionTreeClassifier()
student_model.fit(input_dataset,output_dataset)
Student_grade_point=student_model.predict([[4.0,5,300000,24,69,69]])
Student_grade_point
```
