```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.flipkart.com/search?q=watch+for+men&sid=r18%2Cf13&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=watch+for+men%7CWrist+Watches&requestId=b24d2bcd-7aad-4736-b85e-0fe934841792&as-searchtext=watch%20for%20men"
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
brand=[]
description=[]
price=[]
```

```python
names=soup.find_all('div',class_="syl9yP")
for i in names:
    name=i.text
    brand.append(name)
brand=brand[0:40]   
print(brand)
len(brand)
```

```python
desc=soup.find_all('a',class_="WKTcLC")
for i in desc:
    name=i.text
    description.append(name)
description=description[0:40]
print(description)
len(description)
```

```python
prices=soup.find_all('div',class_="Nx9bqj")
for i in prices:
    name=i.text
    price.append(name)
price=price[0:40]
print(price)
len(price)
```

```python
df=pd.DataFrame({"Brand Name":brand,"Description":description,"Prices":price })
print(df)
```
